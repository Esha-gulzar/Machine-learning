Evaluation matrices (or metrics) in machine learning are quantitative measures used to assess a model’s performance, 
specifically comparing predicted values against actual ground truth.

# MAE (mean absolute error) 
$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$


Components:

Total number of observations.$$n$$ 

 Actual value.$$y_i$$

 Predicted value.$$\hat{y}_i$$

$$|y_i - \hat{y}_i|$$Absolute difference between actual and predicted

## Advantage of Mae
#1.Have same unit as of output column
2.Robust to outliers

## disadvantage of Mae
1.Optimization Issues: The absolute function is not differentiable at zero, which can create difficulties in gradient-based optimization methods.


2.Doesn't Penalize Large Errors: Because it does not square the errors, MAE treats all errors equally regardless of magnitude, which might be undesirable when extreme errors are costly.

## MSE (mean squared error)
The Mean Squared Error (MSE) formula, used to calculate the average squared difference between estimated and actual values, is: 

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$


Actual values $$y_i$$

Predicted values$$\hat{y}_i$$

 Number of data points $$n$$

 Summation from 1 to n $$\sum_{i=1}^{n}$$


## Advantage of Mse
Differentiable and Smooth: MSE provides a smooth gradient, making it easy to calculate and highly efficient for optimization algorithms like gradient descent.

#Penalizes Large Errors: Because it squares the errors, large discrepancies between predicted and actual values have a disproportionately large effect on the loss, forcing the model to prioritize reducing big mistakes.

## disAdvantage of Mse

Sensitive to Outliers: A single extreme outlier can disproportionately inflate the MSE, leading to a model that fits to the noise rather than the data.

Reduced Interpretability: Because the unit of error is the square of the target variable’s unit, it is less intuitive than alternatives like Mean Absolute Error (MAE).

## RMSE root over of MSE 

$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$


## Advantage of RMse
1.Identical Units: Because it is in the same units as the target variable (e.g., dollars, meters), it is easy to interpret the average error directly.

2.Penalizes Large Errors: By squaring the errors before averaging, RMSE penalizes large errors more heavily than small ones, making it ideal when large mistakes are undesirable.

3.Differentiable: Being a continuous and differentiable function, it is often preferred as a loss function for gradient-based optimization in machine learning.

## disAdvantage of RMse

1.Sensitivity to Outliers: RMSE is highly sensitive to outliers because squaring large errors gives them disproportionate weight.


2.Less Robust: If the dataset contains extreme outliers that are not representative, RMSE may misrepresent the true model performance.

## R2 score (coefficient of determination)


 (coefficient of determination) score measures the proportion of variance in the dependent variable explained by a linear regression model. It ranges from 0 to 1 (usually), with higher values indicating better fit,

$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$


Interpretation


$$R^2=1$$Perfect fit; the model explains all variance, and predictions match data exactly.


$$R^2=1$$ The model performs no better than predicting the mean value.
$$R^2<0$$ Negative Indicates the model fits the data worse than a horizontal line representing the mean, often occurring with non-linear models or models lacking an intercept

The 
 score (coefficient of determination) measures how well a regression model explains the variance of the dependent variable compared to a simple mean, with values typically ranging from 0 to 1.

## Advantage of R2 Score

Interpretability: It provides a simple, intuitive percentage (0% to 100%) indicating how much variation is explained.
Model Comparison: Allows for easy comparison of different models on the same data; higher 
 generally indicates a better fit.


Standardization: The score is scale-independent, meaning it works the same regardless of the units of the dependent variable.
Goodness of Fit: It is an excellent quick measure to check if a linear regression model captures the underlying data patterns.

## disAdvantage of R2 Score¶

Overfitting (Doesn't Punish Complexity): 
 $$R^2$$always increases or stays the same when more variables are added, even if they are irrelevant. This can lead to overfitting, where a model performs well on training data but poorly on new data
