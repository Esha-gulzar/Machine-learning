```python
class ManualLR:
    def __init__(self):
        self.m=None
        self.b=None
    def fit(self,X_train,y_train):
        num=0
        dnum=0
        for i in range (X_train.shape[0]): ##We use shape[0] because we want to loop through every row (every data point) one by one
           num=num+(X_train[i]-X_train.mean())*(y_train[i]-y_train.mean())
           dnum=dnum+(X_train[i]-X_train.mean())*(X_train[i]-X_train.mean())
        self.m=num/dnum
        self.b=y_train.mean()-(self.m*X_train.mean())
        print(self.m)
        print(self.b)
    def predict(self,X_test):
        print(X_test)
        return self.m*X_test+self.b
```
