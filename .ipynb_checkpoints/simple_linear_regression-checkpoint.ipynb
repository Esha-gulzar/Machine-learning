## Simple linear regression

Simple linear regression aims to find the "best-fit" straight line, modeled as that minimizes the errors (residuals) between predicted and actual data points. 
By determining optimal slope (m) and intercept (b) values—often using least squares—the model produces more accurate predictions for dependent variables.#unseen datapoint
y=mx+b

## Optimization Techniques:


1.Closed form solution (finding solution by just directly applying formula) this is good for lower dimensions like 1D
Ordinary Least Squares (OLS): Calculates the parameters directly to minimize the sum of squared differences.

2.Non-closed form solution(Gradient Descent: Iteratively adjusts m and b to find the minimum cost.)

## In simple linear regression, the term 

refers to the Residual Sum of Squares (RSS), also commonly called the Sum of Squared Errors (SSE). 
 $$J= di^2 for i=1 to n$$

The variable 
 represents the residual (or error) for the 
i-th data point. It is the vertical distance between the observed value 
 and the value predicted by the regression line
    
  $$SSE = sum of (y_i - y_hat_i)^2 for i=1 to n$$



$$d=y_i - y_hat_i$$
y_hat_i=predicted value

$$yi​=β0​+β1​xi​+ϵi​$$

Consequently, the objective function 
 is defined as:
$$J=i=1∑n​(yi​−y^​i​)2$$

In simple linear regression, where the predicted value is 


, the function becomes:
$$J(β0​,β1​)=i=1∑n​(yi​−(β0​+β1​xi​))2$$


Error function is just like a normal function that depends on it parameter like wise,error function depends on variations in the values of m and b

## 2 possible situations 

## Situation 1 — b=0, m varies:
The line is forced to pass through the origin (0,0). As you increase m, the line rotates like a clock hand. The errors (red lines) shrink or grow depending on how well the slope matches the data. SSE is minimized at the slope that best fits the data.
## Situation 2 — m=0, b varies:
The line becomes perfectly horizontal. As b changes, the whole line shifts up or down. Since the slope is 0, the line can never follow the trend in the data, so SSE stays large no matter what b you pick.
Key insight:


$$ei​$$ changes every time the line moves — either rotating (m changes) or shifting (b changes)
The goal of regression is to find the combination of m and b that makes all those red lines as short as possible

## Minima and maxima
## Global minima and local minima

In machine learning, maxima and minima are the highest and lowest points of a loss function, representing the best (minimum loss) or worst (maximum loss) model performance. The global minimum is the absolute lowest point in the entire function's domain, representing the optimal parameters that yield the lowest possible training error across the entire model space. 

Key Concepts in ML Optimization:
Global Minimum: The lowest possible value of the loss function. It indicates the best possible solution to the learning problem, where the model has achieved the lowest error.
Local Minimum: A point lower than its immediate neighbors but not the lowest in the entire range. Optimization algorithms like Gradient Descent can get "trapped" in these, resulting in sub-optimal models.
Maxima: Points where the function's value is higher than neighbors, usually avoided in minimization tasks.

This makes intuitive sense — a hill rises then falls (maxima), a valley falls then rises (minima).


## In Simple Linear Regression we differentiate to find the Global Minima (not maxima).

## Evaluation Matrices for regression

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print("MAE",mean_squared_error(y_test,y_pred))
print("MSE",mean_absolute_error(y_test,y_pred))
print("RMSE",np.sqrt(mean_absolute_error(y_test,y_pred)))
print("R2 score",r2_score(y_test,y_pred))
```

```python
r2=r2_score(y_test,y_pred)
n=X_test.shape[0]
k=X_train.shape[1]
```

```python
ad_r2=1-((1-r2)*(n-1))/(n-1-k)
print(ad_r2)
```
